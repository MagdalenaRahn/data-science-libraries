```python
roles = pd.DataFrame({
    'id' : [1,2,3,4],
    'name': ['admin','author','editor','commentor']
})

roles
```

```python
# 2 What is the result of using a right join on the DataFrames?

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'right',
            indicator = True
            )

# Takes all the rows of the R table, 
    # and also the rows of the L table where these rows (keys) match the R table

    
    # role_id is foreign key of users table, referencing the 'id' of roles table
```

```python

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'inner',
            indicator = False
            )
```

```python
# 3 What is the result of using an outer join on the DataFrames?

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'outer',
            indicator = True
            )

# outer join returns those with a match in the L and the R tables' rows

# preserves all the values of both tables
```
