```python
# Exercises II

# 1 / Copy the users and roles DataFrames from the examples above.


users = pd.DataFrame({
    'id':[1,2,3,4,5,6],
    'name':['Pablo','Boyan','Tsvetana','Vincent','Faris','Gloria'],
    'role_id':[1,2,3,3,np.nan,np.nan]
    
    
})

users
```

```python
roles = pd.DataFrame({
    'id' : [1,2,3,4],
    'name': ['admin','author','editor','commentor']
})

roles
```

```python
# 2 What is the result of using a right join on the DataFrames?

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'right',
            indicator = True
            )

# Takes all the rows of the R table, 
    # and also the rows of the L table where these rows match the R table
```

```python

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'inner',
            indicator = False
            )
```

```python
# 3 What is the result of using an outer join on the DataFrames?

users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'outer',
            indicator = True
            )

# outer join returns those with a match in the L and the R tables' rows
```

```python
# 4 What happens if you drop the foreign keys from the DataFrames and try to merge them?


(users.merge(roles, 
            left_on='role_id', 
            right_on = 'id', 
            how = 'outer',
            indicator = True
            )
 .drop(columns = ['role_id'])
)
```
